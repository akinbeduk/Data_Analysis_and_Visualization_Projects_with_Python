```python
import numpy as np
import pandas as pd
df=pd.read_csv("country_vaccination_stats.csv")
df.daily_vaccinations=df.groupby("country")["daily_vaccinations"].transform(lambda x: x.fillna(0 if x.sum()==0 else x.min()))
```

**5. Code Implementation Task: Implement code to list the top-3 countries with highest median daily vaccination numbers by considering missing values imputed version of dataset.
Please  provide the link to your code as answer to this question.**

```python
df.groupby("country")["daily_vaccinations"].median().sort_values(ascending=False)[:3]
```
